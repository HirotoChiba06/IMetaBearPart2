```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
#  pytrendsindex_col=0
df = pd.read_csv("../data/trends.csv", index_col=0).reset_index()

#  rename
df = df.rename(columns={"index": "date"})

# isPartial
if "isPartial" in df.columns:
    df = df.drop(columns=["isPartial"])

# 
df["date"] = pd.to_datetime(df["date"])

# 
plt.figure(figsize=(12,6))
plt.plot(df["date"], df["Kyoto travel"], label="Kyoto", linewidth=2)
plt.plot(df["date"], df["Tokyo travel"], label="Tokyo", linewidth=2, linestyle="--")
plt.title("Google Trends: Kyoto vs Tokyo Travel (20152025)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Interest (0100)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
